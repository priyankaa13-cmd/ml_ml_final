```python
sns.scatterplot(x='Selling_Price', y='Year', data=df, color='green')

# Add title and labels
plt.title('selling price vs year')
plt.xlabel('Selling_price')
plt.ylabel('Year')

# Show the plot
plt.show()
```

```python
count_data = df['Fuel_Type'].value_counts()

# Create a bar chart
count_data.plot(kind='bar', color='skyblue')

# Add title and labels
plt.title('fuel types of cars with count')
plt.xlabel('Fuel Tpye')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd




# Initialize the LabelEncoder
label_encoder = LabelEncoder()

df['Fuel_Type'] = label_encoder.fit_transform(df['Fuel_Type'])
df['Seller_Type'] = label_encoder.fit_transform(df['Seller_Type'])
df['Transmission'] = label_encoder.fit_transform(df['Transmission'])


# Display the DataFrame with encoded labels
print(df)
```

```python
sns.pairplot(df)
```

```python
df_numeric = df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven', 'Owner', 'Fuel_Type', 
                 'Seller_Type', 'Transmission']]

# Calculate the correlation matrix
corr_matrix = df_numeric.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```
